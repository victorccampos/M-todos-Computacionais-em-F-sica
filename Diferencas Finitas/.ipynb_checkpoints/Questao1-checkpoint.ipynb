### Questão 1)  
---
 $\boxed{\dfrac{\partial ^2 u}{dt^2} = c^2 \dfrac{\partial ^2 u}{dx^2}}$ 
 
 Utilize o método de **Staggered Leap Frog** para resolver a equação de onda unidimensional assumindo como condição inicial uma onda senoidal, isto é:  

$u(x, t=0) = A sin(kx)$  

com $k = \dfrac{2n\pi}{2L}$ sendo $L$ o comprimento do sistema e $n$ um número inteiro. Explicite as considerações feitas e o conjunto de parâmetros utilizados. Apresente resultados para o perfil da onda y(x, t) e para o perfil de velocidades $v_y = \dfrac{\partial y(x, t)}{\partial t}$


---
$ \text{Os modos que estamos avaliando é}$ $n\pi x / L$

$t = 0$ -> FTCS ;  
$t = 1$ -> Staggered Leap Frog  


$\text{O Staggered Leap Frog a se resolver, na forma de equação de onde é:}$

$\dfrac{u_{j}^{n+1} - 2 u_j^n + u_j^{n-1} }{(\Delta t)^2}$ = $v^2 \dfrac{u_{j+1}^{n} -2u_j^n + u_{j-1}^n}{(\Delta x)^2}$  

$\text{isolando}$ $u_j^{n+1}$ $\text{:}$  

$\boxed{u_j^{n+1} = \dfrac{(\Delta t)^2 v^2}{(\Delta x)^2} (u_{j+1}^n - u_{j-1}^n - 2u_j^n) + 2u_j^n - u_j^{n-1}}$.  

$\text{Na estabilidade da condição de Courant}$, $v\Delta t \le \Delta x$. $\text{Faremos nesse caso}$ $v\Delta t = \Delta x$



```python
#  Parametros Condicao Inicial u(x,0) = A sin(kx) k = n pi / L
A = 1
n = 1
L = 1
k  = (2*n*np.pi) / (2*L)
```

```python
Nx = 100
Nt = 10_000
dx = 1/(Nx-1) # 0.01
dt = 1/(Nt-1) # 0.001
v = 100 # v*dt = dx
x = np.linspace(0, L, Nx)
```

```python
v*dt 
```

```python
u = np.zeros((Nx, Nt)).T
print(u)
print(f"Tamanho de u: {len(u)}\nTamanho de u(x,0): {len(u[0])}\nTipo de u(x,0): {type(u[0])}")
```

```python
# Condição Inicial
plt.title("Condição Inicial")
u[0] = A * np.sin(k*x) ; plt.plot(x, u[0]); ### y(x, t=0)
```

$\text{Para implementar o primeiro passo de tempo fazer um Forward Time Centered Space.}$  

$u_j^{n+1} = u_j^n - \dfrac{v \Delta t}{(2 \Delta x)}(u_{j+1}^n - u_{j-1}^n)$  

$\text{e depois o Staggered Leap Frog:}$

$\boxed{u_j^{n+1} = \dfrac{(\Delta t)^2 v^2}{(\Delta x)^2} (u_{j+1}^n - u_{j-1}^n - 2u_j^n) + 2u_j^n - u_j^{n-1}}$.



```python
def FTCS(u0, dx: float, dt: float, v:int):
    courant = (v*dt/2*dx)
    u_new =  np.zeros_like(u0)
    for j in range(1, Nx-1):
        u_new[j] = u0[j] - courant * (u0[j+1] - u0[j-1])
    return u_new
    
```

```python
def StagLeapFrog(u: np.ndarray, dx, dt, Nx, Nt, v) -> None:
    for n in range(1, Nt-1):
        if n == 1:
            u[n] = FTCS(u[0], v=v, dx=dx, dt=dt)
        for j in range(1, Nx-1):
            u[n+1][j] = (dt*v/dx)**2 * (u[n][j+1] - u[n][j-1] - 2*u[n][j]) + 2*u[n][j] - u[n-1][j]     
```

```python
StagLeapFrog(u, dx, dt, Nx, Nt, v);
```
