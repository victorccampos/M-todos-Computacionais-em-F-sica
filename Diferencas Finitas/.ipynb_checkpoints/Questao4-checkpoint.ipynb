Função de onda inicial

```python
def gauss_packet(x: np.ndarray, x0: float, p: float, omega: float) -> np.ndarray:    
    gaussiana = np.sqrt(1 / np.sqrt(np.pi) / omega) * np.exp(-(x - x0)**2 / 2 / omega**2)
    return gaussiana * np.exp(1j * p * (x-x0))
```
