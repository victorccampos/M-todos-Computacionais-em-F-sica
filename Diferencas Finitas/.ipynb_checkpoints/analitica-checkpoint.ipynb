```python
import numpy as np
import matplotlib.pyplot as plt


# Parâmetros
L = 5
N = 1000
dx = L / (N - 1)
x = np.linspace(0, L, N)

# Condições iniciais
x0 = L / 4
k0 = 2
sigma0 = 0.4

# Constantes físicas
hbar = 1.0  # Constante de Planck reduzida
m = 1/2     # Massa da partícula

# Função de onda analítica
def psi_analytic(x, t):
    sigma = sigma0 * np.sqrt(1 + (hbar * t / (m * sigma0 ** 2)) ** 2)
    return np.exp(1j * k0 * x - (x - x0) ** 2 / (4 * sigma ** 2 + 1j * hbar * t / m)) / np.sqrt(2 * np.pi * sigma ** 2)

# Tempo
tmax = 1
nt = 100
t_values = np.linspace(0, tmax, nt)

# Plot
plt.figure(figsize=(10, 6))
for t in t_values:
    psi = psi_analytic(x, t)
    plt.plot(x, np.abs(psi) ** 2, label=f't = {t:.2f}')
plt.title('Densidade de Probabilidade ao longo do tempo')
plt.xlabel('Posição')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
L = 5
N = 1000
dx = L / (N - 1)
x = np.linspace(0, L, N)

# Condições iniciais
x0 = L / 4
k0 = 2
sigma0 = 0.4

# Constantes físicas
hbar = 1.0  # Constante de Planck reduzida
m = 1.0     # Massa da partícula

# Função de onda analítica
def psi_analytic(x, t):
    sigma = sigma0 * np.sqrt(1 + (hbar * t / (m * sigma0 ** 2)) ** 2)
    x_shift = x0 + (hbar * k0 * t) / m
    return np.exp(1j * k0 * x_shift - (x - x_shift) ** 2 / (4 * sigma ** 2 + 1j * hbar * t / m)) / np.sqrt(2 * np.pi * sigma ** 2)

# Tempo
tmax = 1
nt = 100
t_values = np.linspace(0, tmax, nt)

# Plot
plt.figure(figsize=(10, 6))
for t in t_values:
    psi = psi_analytic(x, t)
    plt.plot(x, np.abs(psi) ** 2, label=f't = {t:.2f}')
plt.title('Densidade de Probabilidade ao longo do tempo')
plt.xlabel('Posição')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.show()
```
