```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros Numéricos
L = 1
Nx = 1_000
Nsteps = 620 # ou menos
x = np.linspace(0, L, Nx)
dx = x[1] - x[0] # 10e-3
dt = 1E-3
alpha = -1j*dt / dx**2  # D = 1
```

```python
def psi0(x: np.ndarray, x0: float, p: float, omega=0.1) -> np.ndarray:
    """
    p: momento da partícula | x0: posição do centro do pacote de onda
    """
    gaussian = np.sqrt(1 / np.sqrt(np.pi) / omega) * np.exp(-(x - x0)**2 / 2 / omega**2)   
    return np.exp(1j*p*(x-x0)) * gaussian
```

```python
psi_t = np.zeros((Nsteps, Nx))
psi_t[0] = psi0(x=x, x0=0.4, p=1.0)
plt.plot(x, psi_t[0])
```

```python
# Matriz tridiagonal do Crank-Nicolson
N = len(x)
A = np.zeros((N-2, N-2), dtype=complex)
B = np.zeros((N-2, N-2), dtype=complex)

# Diagonal Principal
for i in range(0, N-2):
    A[i,i] = 2 + 2*alpha
    B[i,i] = 2 - 2*alpha

# Outras duas diagonais
for i in range(0, N-3):
    A[i+1, i] = -alpha
    A[i, i+1] = -alpha
    B[i+1, i] = alpha
    B[i, i+1] = alpha

Ainv = np.linalg.inv(A)
```

```python
for i in range(Nsteps - 1):
     psi_t[i+1][1:-1] = np.dot(Ainv, np.dot(B, psi_t[i][1:-1]))
```

```python
import matplotlib.animation as animation
 # tamanho das figuras
font = {'family': 'serif',
        'weight': 'normal',
        }
# Função para atualizar o gráfico a cada quadro da animação
def update_plot(i, data, line):
    line.set_ydata(data[i])
    return line,

# Configuração inicial do gráfico
fig, ax = plt.subplots()
plt.suptitle("Evolução Temporal", font=font)
plt.ylabel('psi', font=font)
plt.xlabel('x', font=font)
plt.tight_layout()
line, = ax.plot(x, psi_t[0], c='k')

# Criação da animação
ani = animation.FuncAnimation(fig, update_plot, frames=len(psi_t), fargs=(psi_t, line), interval=20)

# Salvar a animação em um arquivo de vídeo
ani.save('TESTE.mp4', writer='ffmpeg')

# Mostrar o gráfico
plt.show()
```
