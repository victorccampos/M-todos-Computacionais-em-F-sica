```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 4] # tamanho das figuras
```

$y_{n+1} = r y_n(1-y_n)$

```python
def logistic(y0, r, nmax):
    y_list = [y0]
    for n in range(1, nmax + 1):
        y_list.append(r * y_list[n-1] * ( 1 - y_list[n-1]) )
    return y_list
```

#### Questão 1: Faça gráficos de $y_n$ como função de n para $r=0.5$. Use $y_0 = 0.5$ e $n$ entre 1 e 50.

```python
rs = [0.5, 2.5, 3.1, 3.5, 3.7]
simulacoes = []
for r in rs:
    simulacao = logistic(y0=0.5, r=r, nmax=50)
    simulacoes.append(simulacao)
```

```python
# %matplotlib qt
plt.figure(figsize=(12,6))
for i in range(len(rs)):
    plt.plot(simulacoes[i], label=f'{rs[i]}')
    plt.xlabel('n')
    plt.ylabel('y')
    plt.legend(loc='lower right')
    plt.title("n = 50 " + r"$y0$ = 0.5")
    plt.grid()
```

#### Refaça os gráficos anteriores considerando três condições iniciais consideravelmente diferentes, ou seja, trace em um mesmo gráfico as curvas para $y_0 = 0.25$, $y_0 = 0.5$ e $y_0 = 0.75$. Comente

```python
initial_conditions = [0.25, 0.5, 0.75]
rs = [0.5, 2.5, 3.1, 3.5, 3.7]
for r in rs:
    plt.show()
    for initial_condition in initial_conditions:
        # print(r)
        # print(initial_condition)
        # print()
        plt.plot(logistic(r=r, y0=initial_condition, nmax=50), label=f'y0 = {initial_condition}')
        plt.legend()
        plt.title(f'r = {r}')
        plt.xlabel('n')
        plt.ylabel('População y')
```

```python
initial_conditions = [0.5, 0.501, 0.5001]
rs = [0.5, 2.5, 3.1, 3.5, 3.7]
for r in rs:
    plt.show()
    for initial_condition in initial_conditions:
        # print(r)
        # print(initial_condition)
        # print()
        plt.plot(logistic(r=r, y0=initial_condition, nmax=50), label=f'y0 = {initial_condition}')
        plt.legend()
        plt.title(f'r = {r}')
        plt.xlabel('n')
        plt.ylabel('População y')
```

#### Considere 10000 valores de r igualmente espaçados entre $10^{-5}$ e 4. Para cada um desses valores de r, vamos iterar a equação logística 1000 vezes,  i.e., nmax = 1000.  
#### Em um gráfico de $y_n$ como função de r, vamos plotar para cada valor de r os 100 últimos passos da evolução. Teremos 100 pontos no eixo $y$ do gráfico para cada valor de $r$, que estará no eixo $x$. 
#### A figura gerada é conhecida como diagrama de bifurcação. Essa figura traz inúmeras informações importantes. Tire um tempo para pensar nela e analisar o que encontrou.

```python
rs = np.linspace(0.00001, 4, 10_000)
Y_LIST = []
for r in rs:
    y_r =logistic(r=r, y0=0.5, nmax=1000)
    Y_LIST.append(y_r[-100:])
```

```python
%matplotlib qt
plt.style.use("dark_background")
font = {'family': 'serif', 'weight': 'normal', 'size': 20}
plt.title("Diagrama de bifurcação", font=font)
plt.plot(rs[:], Y_LIST[:], '.', markersize=1)
plt.xlabel('r');
plt.ylabel('População y');
plt.annotate("Autor: João Victor Campos" + "\nData:13/03/2024", xy=[0,0.9])
plt.annotate(r"$y_{n+1} = ry_n(1-y_n)$", xy=[2.0,0.9]);
```
